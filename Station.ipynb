##**Station Object Part**

```python
class Station:
  def __init__(self, name, lines, latitude, longtitude, around_stations):
    self.name = name
    self.lines = lines
    self.latitude = latitude
    self.longtitude = longtitude
    self.around_stations = around_stations #dictionary
    
  def print(self):
    print("name: " + self.name)
    print("lines: ", self.lines)
    print("latitude: " + str(self.latitude))
    print("longtitue: " + str(self.longtitude))
    print("around_stations: " ,end='')
    print(self.around_stations)
  def returnlines(self):
    return self.lines
  
```

# **JSON Parsing + Construct Stations Object Part**

```python
import json

with open('capitalStations.json',encoding ='utf-8-sig')as json_file:
  json_data = json.load(json_file)


station_info = {} #info저장
station_list=[]   #전체 Station 저장



for a in json_data:
    station_info[a["name"]]=a["lines"]

for _ in json_data:
 
  if(len(_["around_stations"])%2==0): # around_stations에 겹치는 역이 없을 때,

    tmp_around = _["around_stations"] #around_stations 임시 저장
    tmp_dict={}                       #최종 딕셔너리 생성

    for i in _["lines"]:              #line 별로 딕셔너리 생성 ex) { "01호선" : ["가능","양주"] }
                                      
      tmp_list=[]                     #tmp_list 생성 ex) ["가능","양주"]

      for j in range(2):              #겹치는 역이 없을 때는, 이전역 다음역이 전부 다르다.
                                      
        try:
          tmp_list.append(tmp_around[-1])   #tmp_list에 마지막 around 역부터 2개씩 넣어줌.
          del tmp_around[-1]
        except:
          pass
          
      tmp_dict[i]=tmp_list            #최종, 딕셔너리 생성

    station_list.append(Station(_["name"], _["lines"], _["latitude"], _["longitude"], tmp_dict))  #위의 값들을 토대로 Station 객체 생성.
    

  else:                           # around_stations에 겹치는 역이 있을 때,
    
      tmp_dict = {}
      tmp_list = []
      for i in _["around_stations"]:

        tmp_list=station_info[i]


        for k in tmp_list:
          if k in _["lines"]:
        
            if(tmp_dict.get(k)==None):
              tmp_listing=[]
              tmp_listing.append(i)
              tmp_dict[k]=tmp_listing
            else:            
              a=tmp_dict[k]
              a.append(i)
              tmp_dict[k]=a
               
      station_list.append(Station(_["name"], _["lines"], _["latitude"], _["longitude"], tmp_dict))
   

trash = ["경강선","경춘선","김포도시철도","서해선","용인경전철","우이신설경전철","의정부경전철","인천2호선","자기부상철도","인천선"]



for a in station_list:

  a.print()
  print("")

print(len(station_list))



```
